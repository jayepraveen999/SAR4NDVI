## Plotting the NDVI graph of one random pixel value for all the cloud cover image over 2019-2020

### Importing necessary packages

```python
import numpy as np
import pandas as pd
import altair as alt
```

### Reading data

```python
data = pd.read_csv('ndvi_seq_2.csv')
# data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
```

```python
data['NDVI Sentinel 2'] = data['ndvi']
data['NDVI DL Model'] = data['ndvi_pred']
data['Index'] = data['Unnamed: 0']
data['Date 2019-2020'] = data['time']
data = data.drop(columns=['ndvi','ndvi_pred','Unnamed: 0','time'],axis=1)
data
```

### Plotting the graphs using altair: Better Visualization

```python
brush = alt.selection_interval(encodings= ["x"])


# plotting the dataframe using altair


hist1 = alt.Chart(data).mark_bar().encode(x = "Index",
    color=alt.condition(brush, alt.value("cadetblue"), alt.value('lightgray'))
).interactive().properties( width=800,title = 'NDVI Value Time Series').add_selection(brush)

hist1.configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray'
)


# concatenating all plots and making them interactive

hist1.encode(alt.Y("NDVI Sentinel 2",scale=alt.Scale(domain=[-0.5, 0.9]),title = 'NDVI Value: Sentinel 2 [-1 to +1]'),color=alt.condition(brush, alt.value("cadetblue"), alt.value('lightgray')), tooltip=["Index", "NDVI Sentinel 2"]
) & hist1.encode(alt.Y("NDVI DL Model",title = 'NDVI Value: DL Model [-1 to +1]'),color=alt.condition(brush, alt.value("cadetblue"), alt.value('lightgray')), tooltip=["Index", "NDVI DL Model"]
)
```
